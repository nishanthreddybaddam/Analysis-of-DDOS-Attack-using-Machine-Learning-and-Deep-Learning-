```python
from pycaret.classification import *
import pandas as pd
data = pd.read_csv('kdd_train.csv')
dt = setup(data = data, target = 'labels')
```

```python
print(data['labels'].value_counts())
```

```python
from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
```

```python
#dropping missing values
data = data.dropna(subset=['labels'])

# init setup on exp
exp.setup(data, target = 'labels')
```

```python
# compare baseline models
best_model = compare_models()
```

```python
# plot confusion matrix
plot_model(best_model, plot = 'confusion_matrix')
```

```python
plot_model(best_model, plot = 'auc')
```

```python
plot_model(best_model, plot = 'feature')
```

```python
evaluate_model(best_model)
```

```python
holdout_pred = predict_model(best_model)
```

```python
# Plot Decision Boundary (for models that support it)
plot_model(best_model, plot='boundary')
```

```python
# Plot Class Prediction Error
plot_model(best_model, plot='error')
```

```python
# Plot learning curve
plot_model(best_model, plot='learning')
```

```python
# Plot Validation Curve
plot_model(best_model, plot='vc')
```

```python
# Plot Class Distribution
plot_model(best_model, plot='class_report')
```
